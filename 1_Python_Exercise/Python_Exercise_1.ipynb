In this exercise you will be implementing a function calculating the Fibnoacci series. The characteristics of a Fibonacci series is that each element is the sum of the previous two elements. In mathemtical terms it is defined as follows:  
f(0) = 0  
f(1) = 1  
f(n) = f(n-1) + f(n-2); n>1

```python
import matplotlib.pyplot as plt # Library for ploting data later
import numpy as np
```

```python
def fib(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return fib(x-1) + fib(x-2)
```

The following code checks, if your implementation is correct. If it runs without throwing an error, your implementation is likely correct.  
It does so by using the 'assert' statement. In python, assert followed by a False (boolean) will throw an error

```python
try:
    assert fib(0)  == 0
    assert fib(1)  == 1
    assert fib(2)  == 1
    assert fib(10) == 55
    print('Your implementation seems to be correct. Well done!')
except:
    print('There is an error in your implementation.')
```

```python
# Create a list of the first 20 fibonacci numbers
fib_list = [fib(z) for z in range(20)]
```

```python
# Test, if your fib_list really has 20 elements
assert len(fib_list) == 20
```

```python
# Plot the fibonacci numbers
plt.figure(figsize=(10,10))
plt.plot(range(20),fib_list,'rx') # r = red, x = Creates an x where the data points are
plt.grid(True) # Show a grid
plt.xlabel('X')
plt.ylabel('Fibonacci(X)')
plt.show()
```
